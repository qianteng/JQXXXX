```python
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
train = pd.read_csv(
    '~/Kaggle/walmart/train.csv',
    dtype={'TripType': str},
)
```

```python
scan_count = train.groupby(['TripType', 'VisitNumber'])['ScanCount'].sum()
scan_count[scan_count < 0].reset_index()
```
