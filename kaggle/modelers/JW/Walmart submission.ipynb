```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
rawcsv = pd.read_csv("Desktop/Walmart/train.csv",dtype={'Upc':object})
```

```python
rawcsv_test = pd.read_csv("Desktop/Walmart/test.csv",dtype={'Upc':object})
```

```python
weekDays = rawcsv['Weekday'].unique()
```

```python
weekDayDummy = {q : (rawcsv['Weekday']==q)*1 for q in weekDays} 
weekDayDummy_test = {q : (rawcsv_test['Weekday']==q)*1 for q in weekDays}# make weekdays as dummy variables
```

```python
wkData = pd.DataFrame(weekDayDummy, columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday','Sunday'])
wkData_test = pd.DataFrame(weekDayDummy_test, columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday','Sunday'])
```

```python
dept = rawcsv['DepartmentDescription'].unique() # code DepartmentDescription as dummy variable
dept = pd.DataFrame(dept)
```

```python
dept = dept.fillna('NaN')
rawcsv['DepartmentDescription'] = rawcsv['DepartmentDescription'].fillna('NaN')
rawcsv_test['DepartmentDescription'] = rawcsv_test['DepartmentDescription'].fillna('NaN')
```

```python
dict1 = {q:rawcsv['DepartmentDescription']==q for q in dept.ix[:,0]}  
dict1_test = {q:rawcsv_test['DepartmentDescription']==q for q in dept.ix[:,0]}  

```

```python
dicData = pd.DataFrame(dict1)
dicData = dicData*1
dicData_test = pd.DataFrame(dict1_test)
dicData_test = dicData_test*1
```

```python
mergedData = pd.merge(wkData, dicData, left_index=True, right_index=True)
mergedData = pd.merge(rawcsv, mergedData, left_index=True, right_index=True)
mergedData = mergedData.drop('Weekday', axis=1)
mergedData = mergedData.drop('Upc', axis = 1)
mergedData = mergedData.drop('FinelineNumber', axis = 1)
mergedData = mergedData.drop('DepartmentDescription', axis = 1)
```

```python
mergedData_test = pd.merge(wkData_test, dicData_test, left_index=True, right_index=True)
mergedData_test = pd.merge(rawcsv_test, mergedData_test, left_index=True, right_index=True)
mergedData_test = mergedData_test.drop('Weekday', axis=1)
mergedData_test = mergedData_test.drop('Upc', axis = 1)
mergedData_test = mergedData_test.drop('FinelineNumber', axis = 1)
mergedData_test = mergedData_test.drop('DepartmentDescription', axis = 1)
```

```python
mergedData_test
```

```python
for i in range(10,78):
    mergedData.ix[:,i] = mergedData.ix[:,2]*mergedData.ix[:,i] # compute dot product of each dummy varable for dept description with scan count

for i in range(9,77):
    mergedData_test.ix[:,i] = mergedData_test.ix[:,1]*mergedData_test.ix[:,i] # compute dot product of each dummy varable for dept description with scan count

```

```python
mergedData = mergedData.drop('ScanCount', axis=1)
mergedData_test = mergedData_test.drop('ScanCount', axis=1)
```

```python
grouped2 = mergedData.groupby(["VisitNumber"])
grouped2_test = mergedData_test.groupby(["VisitNumber"])
```

```python
temp = grouped2.mean().ix[:,:8] # aggregate by visit number then take mean on the weekday dummy variables
temp_test = grouped2_test.mean().ix[:,:7] # aggregate by visit number then take mean on the weekday dummy variables
```
